# **Linear Regression**

**Definition -** Linear Regression is the supervised Machine Learning model in which the model finds the best fit linear line between the independent and dependent variable i.e it finds the linear relationship between the dependent and independent variable.

**Types**:
- Simple Linear Regresion: where only one independent variable is present and the model has to find the linear relationship of it with the dependent variable
- Multiple Linear Regression: where more than one independent variables for the model to find the relationship.

$$$$



## **Simple Linear Regression**

Mathematically, simple linear regression is defined as

$$
y=\beta_{0}+\beta_{1} x
$$

where, $x$ is an independent variable or predictor variable and $y$ is the dependent variable or response variable.

$$$$

- $\beta_{1}$: is the slope of the line. A value close to zero indicates no relationship with the response variable. Whereas, a large positive or negative value indicates a strong positive or negative relationship with the response variable.

- $\beta_{0}$: is the intercept of the line i.e is a point where the regression model touches the Y-axis. 


$$$$

In order to actually fit a line, we’ll start with a way to quantify how good a line is. 

We’ll then use this to fit the “best” line we can. One way to quantify a line’s “goodness” is to propose a probabilistic model that generates data from lines. Then the “best” line is the one for which data generated from the line is “most likely”. 

This is a commonly used technique in statistics: proposing a probabilistic model and using the probability of data to evaluate how good a particular model is.

#### **Probabilistic Model for Linearly Related Data**

We observe paired data points $\left(x_{1}, y_{1}\right),\left(x_{2}, y_{2}\right), \ldots,\left(x_{n}, y_{n}\right)$, where we assume that as a function of $x_{i}$, each $y_{i}$ is generated by using some true underlying line $y=\beta_{0}+\beta_{1} x$ that we evaluate at $x_{i}$, and then adding some Gaussian noise. Formally,

$$$$
$$
y_{i}=\beta_{0}+\beta_{1} x_{i}+\varepsilon_{i}
$$

$$$$

Here, the noise $\varepsilon_{i}$ represents the fact that our data won't fit the model perfectly. We'll model $\varepsilon_{i}$ as being Gaussian: $\varepsilon \sim \mathcal{N}\left(0, \sigma^{2}\right)$. Note that the intercept $\beta_{0}$, the slope $\beta_{1}$, and the noise variance $\sigma^{2}$ are all treated as fixed (i.e., deterministic) but unknown quantities.

### **Solving for the fit: least-squares regression**

Assuming that this is actually how the data $\left(x_{1}, y_{1}\right), \ldots,\left(x_{n}, y_{n}\right)$ we observe are generated, then it turns out that we can find the line for which the probability of the data is highest by solving the following optimization problem 3 :
$$
\min _{\beta_{0}, \beta_{1}}: \sum_{i=1}^{n}\left[y_{i}-\left(\beta_{0}+\beta_{1} x_{i}\right)\right]^{2}
$$
where $\min _{\beta_{0}, \beta_{1}}$ meqns "minimize over $\beta_{0}$ and $\beta_{1}$ ". This is known as the least-squares linear regression problem. Given a set of points, the solution is:
$$
\begin{aligned}
\hat{\beta}_{1} &=\frac{\sum_{i=1}^{n} x_{i} y_{i}-\frac{1}{n} \sum_{i=1}^{n} x_{i} \sum_{i=1}^{n} y_{i}}{\sum_{i=1}^{n} x_{i}^{2}-\frac{1}{n}\left(\sum_{i=1}^{n} x_{i}\right)^{2}} \\
&=r \frac{s_{y}}{s_{x}} \\
\hat{\beta}_{0} &=\bar{y}-\hat{\beta}_{1} \bar{x}
\end{aligned}
$$

where $\bar{x}, \bar{y}, s_{x}$ and $s_{y}$ are the sample means and standard deviations for $x$ values and $y$ values, respectively, and $r$ is the correlation coefficient, defined as
$$
r=\frac{1}{n-1} \sum_{i=1}^{n}\left(\frac{x_{i}-\bar{x}}{s_{x}}\right)\left(\frac{y_{i}-\bar{y}}{s_{y}}\right)
$$

By examining the second equation for the estimated slope $\hat{\beta}_{1}$, we see that since sample standard deviations $s_{x}$ and $s_{y}$ are positive quantities, the correlation coefficient $r$, which is always between $-1$ and 1 , measures how much $x$ is related to $y$ and whether the trend is positive or negative.

### **Test and Intervals**

- **Slope**: For the slope $\beta_{1}$, our test statistic is
$$
t_{\beta_{1}}=\frac{\hat{\beta}_{1}-\beta_{1}}{s_{\beta_{1}}},
$$
which has a Student's $t$ distribution with $n-2$ degrees of freedom. The standard error of the slope $s_{\beta_{1}}$ is
and the mean squared error $\hat{\sigma}^{2}$ is
$$
\hat{\sigma}^{2}=\frac{\overbrace{\sum_{i=1}^{n}\left(y_{i}-y_{i}\right)^{2}}^{\text {how large the errors are }}}{n-2}
$$
These terms make intuitive sense: if the $x$-values are all really close together, it's harder to fit a line. This will also make our standard error $s_{\beta_{1}}$ larger, so we'll be less confident about our slope. The standard error also gets larger as the errors grow, as we should expect it to: larger errors should indicate a worse fit.

- **Intercept**: For the slope $\beta_{0}$, our test statistic is
$$
t_{\beta_{1}}=\frac{\hat{\beta}_{0}-\beta_{0}}{s_{\beta_{0}}},
$$ 
which is also $t$-distributed with $n-2$ degrees of freedom.
$$$$
The standard error is
$$
s_{\beta_{0}}=\hat{\sigma} \sqrt{\frac{1}{n}+\frac{\bar{x}^{2}}{\sum_{i}\left(x_{i}-\bar{x}\right)^{2}}},
$$
and $\hat{\sigma}$ is given by Equation (3.9).

- **Correlation**: For the correlation coefficient $r$, our test statistic is the standardized correlation
$$$$
$$
t_{r}=r \sqrt{\frac{n-2}{1-r^{2}}}
$$ 
$$$$
which is $t$-distributed with $n-2$ degrees of freedom. Figure $3.4$ plots $t_{r}$ against $r$.

- **Prediction**: Let's look at the prediction at a particular value $x^{*}$, which we'll call $\hat{y}\left(x^{*}\right)$. In particular:
$$$$
$$
\hat{y}\left(x^{*}\right)=\hat{\beta}_{0}+\hat{\beta}_{1} x^{*} .
$$
$$$$
We can do this even if $x^{*}$ wasn't in our original dataset.
Let's introduce some notation that will help us distinguish between predicting the line versus predicting a particular point generated from the model. From the probabilistic model given by Equation (3.1), we can similarly write how $y$ is generated for the new point $x^{*}$ :
$$
y\left(x^{*}\right)=\underbrace{\beta_{0}+\beta_{1} x^{*}}_{\text {defined as } \mu\left(x^{*}\right)}+\varepsilon_{,}
$$
where $\varepsilon \sim \mathcal{N}\left(0, \sigma^{2}\right)$.
